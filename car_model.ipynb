```python
import pandas as pd

df = pd.read_csv('./car_complain.csv')
df
```

```python
# # 0代表没有,1代表有
df=df.drop('problem',axis = 1).join(df.problem.str.get_dummies(','))
df
```

```python
#车型投诉总数最多
result = df.groupby(['car_model'])['id'].agg(['count'])
result.sort_values('count',ascending=False)
```
